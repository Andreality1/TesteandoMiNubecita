```python
def media(array):
  
  array_move = np.array([])
  array_variable = np.append(array_move, array)

  array_move = array_variable

  int_lenArray = len(array_move)

  float_move = 0.0

  for i in array_move:
    float_move += i

  
  return float_move/int_lenArray
#Que composicion determinamos ahora mismo
#Que ruta yo puedo determinar.
#Que expresion yo puedo sintetizar
#El movimiento que me produce mente.

array_probe_1 = np.array([1.4,1.567,2.3425,1.23415,1.4567])

float_media_probe_1 = media(array_probe_1)
```

```python
#Que clase de aspecto quiero yo ver hecho realidad
#Que movimiento yo quiero dise√±ar
#Que aspecto yo quiero surtir.
def varianzaPower_array(array):
  
  float_media = media(array)
  
  array_move = np.array([])

  array_objeto = np.append(array, array_move)

  array_move = array_objeto

  int_lenArray = len(array_move)

  float_varianza = 0.0

  for i in array_move:
    
    float_varianza +=  (i - float_media) ** 2

  float_varianza = float_varianza/(int_lenArray - 1)

  return float_varianza  

  






array_probe_1 = np.array([1.4,1.567,2.3425,1.23415,1.4567])

float_media_probe_1 = varianzaPower_array(array_probe_1)

print(float_media_probe_1)
```
