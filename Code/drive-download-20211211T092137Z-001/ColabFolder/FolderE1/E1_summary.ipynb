````python
def factorial(num):
  
  iteracion = num

  int_inicio = 1

  int_copia = 1

  for i in range(num):

    int_copia *= int_inicio 
    int_inicio += 1

  return int_copia
````

````python
def combinatoria(n , r):
  int_n = int(n)
  int_r = int(r)

  int_objeto = int_n - 1

  for i in range(r-1):
    int_n *= int_objeto
    #print(int_n)
    int_objeto -= 1 

  int_fact = factorial(int_r)
  C = int_n / int_fact

  return C

````

````python
def guass_normal_float(x):

  float_exp = np.exp((-1/2)*np.power(x,2))

  float_divition  = 1 / (np.sqrt(2 * np.pi))

  float_out = (float_exp * float_divition)

  return float_out
````
